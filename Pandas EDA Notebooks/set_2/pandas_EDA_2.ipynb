```python
# A. Create age bins
bins = [17, 30, 45, 60, float('inf')] # edges of each bin
labels = ["18-30", "31-45", "46-60", "61+"] # readable bin names
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)

# B. Group by age group and calculate average number of orders
df.groupby("AgeGroup", observed=True)["NumOrders"].mean()
```

```python
df["SignupDate"] = pd.to_datetime(df["SignupDate"])
```
